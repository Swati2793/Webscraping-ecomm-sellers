d) Finding Product details and saving in data frame

```python
# lists having files from sponsored and non-sponsored  folder 
files_sponsored =  os.listdir("./sponsored")
files_not_sponsored =  os.listdir("./non-sponsored")
#creating empty lists
Sponsored = []
NonSponsored = []
```

CREATING DATAFRAME FOR SPONSORED ITEMS

```python
Sponsored = []

for file in tqdm(files_sponsored):
    try:
    #### reading html files ###
        htmlFile = open("./sponsored/"+file, "rb")
        fileText = htmlFile.read()
        htmlFile.close
    
    #### product details ###
    
        ## Item Id  - from split filenames
        itemId = int((file.split("."))[0])
        soup = bs(fileText, 'html.parser')
       
        ## Seller Name (selector: span with class mbg-nw) 
        sellerName = (soup.find('span', attrs = {'class':'mbg-nw'})).text
        
        ## Seller Score - (selector: span with class mbg-l)
        sellerScore = int(((soup.find('span', {'class':'mbg-l'})).find('a')).text)
        
        ## Item price - (selector : span with itemprop - price);here ids are different for bidding,discount or original price
        itemPrice = float((soup.find('span', {'itemprop':'price'}))['content'])
        
        ## Number of items sold - (selector : find div with id = main content) and then find number,
        # alternatively , child of span class soup.find('div', {'id':'mainContent'})
        
        #item_sold_div = soup.find('div', {'id':'mainContent'})
        if (re.search("(\d+) sold",(soup.find('div', {'id':'mainContent'}).text))):
            n = re.search("(\d+) sold",(soup.find('div', {'id':'mainContent'}).text))
            itemsSold = int(n.groups()[0])
        else:
            itemsSold = None
        
        ## Best offer available (Selector : form with name viactiondetails)
        bestOfferTag = soup.find('form', {'name':'viactiondetails'})
        if re.search("Best Offer",(bestOfferTag.text)):
            bestOffer = 1
        else:
            bestOffer = 0
            
        ## Title (selector : heading with id ItemTitle)
        title = (soup.find('h1', {'id':'itemTitle'})).find(text=True, recursive=False)
        
        ## Returns allowed (selector: div with id why2buy)
        returnsAllowed = soup.find('div', {'id':'why2buy'})
        if re.search("Returns accepted",(returnsAllowed.text)):
            returnsAccepted = 1
        else:
            returnsAccepted = 0
            
        ## Shipping price (Selector: span with Id = shSummary,0 if free-shipping,None if  )
        shippingPriceParent = soup.find('span', {'id':'shSummary'})
        if re.search('free', shippingPriceParent.text, flags=re.IGNORECASE):
            shippingPrice = float(0)
        elif re.search(r'\$[\d.]+', shippingPriceParent.text):
            shipingPriceTag = re.search("\$([\d.]+)",(shippingPriceParent.text))
            shippingPrice = float(shipingPriceTag.groups()[0])
        else:
            shippingPrice = None
            
        ## Condition (Selector: div with id = vi-itm-cond)
        
        itemCondition = (soup.find('div', {'id':'vi-itm-cond'})).text
        
        ## Sponsored or Non-Sponsored
        sponsored_yn = "Y"
        
        ## Creating final dataframe with eBay Data
        
        ebay_data = [itemId,             
                sellerName,
                sellerScore,
                itemPrice,
                itemsSold,
                bestOffer,
                title,
                returnsAccepted,
                shippingPrice,
                itemCondition,
                sponsored_yn]
        Sponsored.append(ebay_data)
    except:
        continue
```

```python
import mysql.connector
from mysql.connector import errorcode
Mt_cnx = mysql.connector.connect(user='root',
                              host='127.0.0.1')
cursor = Mt_cnx.cursor()

#creating eBay Database
DB_NAME = 'eBay'  

##Capturing Errors
def create_database(cursor):
    try:
        cursor.execute(
            "CREATE DATABASE {} DEFAULT CHARACTER SET 'utf8'".format(DB_NAME))
    except mysql.connector.Error as err:
        print("Failed creating database: {}".format(err))
        exit(1)

## First check if the database exists
try:
    cursor.execute("USE {}".format(DB_NAME))
# If not, call the create_database function
except mysql.connector.Error as err:
    print("Database {} does not exists.".format(DB_NAME))
    if err.errno == errorcode.ER_BAD_DB_ERROR:
        create_database(cursor)
        print("Database {} created successfully.".format(DB_NAME))
        Mt_cnx.database = DB_NAME
    else:
        print(err)
        exit(1)

        
def createTable(TABLE_NAME, TABLE_DESC):
    try:
        print("Creating table {}: ".format(TABLE_NAME))
        cursor.execute(TABLE_DESC)
    except mysql.connector.Error as err:
        if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
            print("already exists.")
        else:
            print(err.msg)
    else:
        print("OK")

TABLE_NAME = 'eBay_items'
TABLE_DESC = (
    "CREATE TABLE `eBay_items` ("
    "  `itemId` VARCHAR(255) NOT NULL,"
    "  `sellerName` VARCHAR(255),"
    "  `sellerScore` int ,"  
    "  `itemPrice` INT ,"
    "  `itemsSold` INT ,"
    "  `bestOffer` INT ,"
    "  `title` VARCHAR(255) ,"
    " `returnsAccepted` INT ,"
    " `shippingPrice` INT ,"
    " `itemCondition` VARCHAR(255) ,"
    " `sponsored_yn` VARCHAR(255)"
 #   "  PRIMARY KEY (`itemId`)"
    ") ENGINE=InnoDB")
createTable(TABLE_NAME, TABLE_DESC)


```
